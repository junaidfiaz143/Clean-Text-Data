```
table = str.maketrans('', '', string.punctuation)
stripped_words = [word.translate(table) for word in words]
print(stripped_words[:100])
```

```
stop_words = set(stopwords.words('english'))
print(stop_words)
```

**Now perform all data cleaning with NLTK**

```
#load dataset
filename = "metamorphosis_clean.txt"
file = open(filename, mode = "rt", encoding = "utf-8-sig")
text = file.read()
file.close()

#tokenize text (split text by words)
tokenize_words = word_tokenize(text)
words = [word.lower() for word in tokenize_words]

#remove punctuations
table = str.maketrans('', '', string.punctuation)
stripped_words = [word.translate(table) for word in words]

#remove other non-alphabetic words
alpha_words = [word for word in stripped_words if word.isalpha()]

#cleaned text data
final_words = [word for word in alpha_words if not word in stop_words]

print(final_words)
```

**Stemming**

```
porter = PorterStemmer()

stemmed_words = [porter.stem(word) for word in final_words]

print(stemmed_words)
```
